```python
# Importing enviroment and packages to be use for out analysis 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("dailyActivity_merged.csv")
print(data.head())
```

```python
#checking for null values
print(data.isnull().sum())
```

```python
#no null values. 
```

```python
print(data.info())
```

```python
data.head()
```

```python
#Changing column ActivityDate from object to datetime64
data['ActivityDate'] = pd.to_datetime(data['ActivityDate'], format="%m/%d/%Y")
```

```python
data['ActivityDate'].head(10)
```

```python
#Creating day of the week out of column ActivityDate 
data["DayOfTheWeek"] = data["ActivityDate"].dt.day_name()
data["DayOfTheWeek"].head(10)
```

```python
data.info()
```

```python
#Creating new column for the total minutes by adding = VeryActiveMinutes,FairlyActiveMinutes,LightlyActiveMinutes,SedentaryMinutes
data["TotalMinutes"] = data["VeryActiveMinutes"] + data["FairlyActiveMinutes"] + data["LightlyActiveMinutes"] + data["SedentaryMinutes"]
data["TotalMinutes"].head()
```

```python
data.info()
```

```python
#Creating new columb for Total Hour 

data["TotalHours"] = round(data["TotalMinutes"] / 60)
data["TotalHours"].head(5)
```

```python
data.describe()
```

```python
plt.figure(figsize=(6,4))
plt.hist(data.DayOfTheWeek,bins=7,width = 0.5, color = "lightpink", edgecolor = "lightskyblue")
plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.title("User logged in frequency across the week")
plt.grid(True)
plt.show()
```

The Middle of the weekday has the highest frequency compared to the weekends and Monday.
One of the things that can be assumed is that users were not exercising as much or as active during the weekends compared to the weekdays.
